#### Proportional Sampling example

```python
from random import uniform
import pandas as pd
# write your python code here
# you can take the above example as sample input for your program to test
# it should work for any general input try not to hard code for only given input examples

#video link for the topic - https://www.appliedaicourse.com/lecture/11/applied-machine-learning-online-course/2876/proportional-sampling/2/module-2-data-science-exploratory-data-analysis-and-data-visualization
# you can free to change all these codes/structure

def pick_a_number_from_list(A):
    # your code here for picking an element from with the probability propotional to its magnitude
    sum=0
    A_dash=[]
    for i in range(len(A)):
        sum+=A[i]
    #print(sum)
    
    #Normalizing using the Sum
    for i in range(len(A)):
        temp=A[i]/sum
        A_dash.append(temp)
    #print(A_dash)           #Normalized list of values     
    
    #Cumulative sum of normalized values
    series = pd.Series(A_dash)
    cumsum = series.cumsum(skipna=False)
    #print("\n\n")
    #print(cumsum)
    
    r=uniform(0.0,1.0)
    #print("r values is : ",r)
    
    #picking a number from the list
    for index in range(len(cumsum)):
        if (r>=cumsum[index] and r<cumsum[index+1]):
            return A[index+1]

        

        
# Function to perform sampling based on it's magnitude
def sampling_based_on_magnitued():
    A = [0,5,27,6,13,28,100,45,10,79]
    #A = [1, 5, 27, 6, 13, 28, 100, 45, 10, 79]
    d = dict()
    #A.sort()
    print("List of elements : ",A,"Sum = ",sum(A))
    for i in range(0,100):
        number = pick_a_number_from_list(A)
        print(number)
        if number not in d:
            d[number] = 1
        else:
            d[number]+=1
    
    print(d)

sampling_based_on_magnitued()
```
