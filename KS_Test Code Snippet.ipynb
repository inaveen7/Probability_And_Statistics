# K-S Test

```python
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
# import warnings
# warnings.filterwarnings("ignore", category=DeprecationWarning) 

#generate a gaussian r.v X
x = stats.norm.rvs(size=1000);  # generating 1000 samples from 'norm':- normal distrubution-->X~N(0,1)
print(len(x),type(x))
sns.set_style('whitegrid')
sns.kdeplot(x, bw=0.5)
plt.show()
```

#### Note:-
-> The above figure looks like Gaussian or Normal distribution with mean=0 and variance=1

```python
stats.kstest(x, 'norm')
```

#### Note:-
-> In the above output, statistic = D and pvalue=p-value what we have seen in the course video.

-> Here, 'pvalue' is quite high. Hence we can say that our random variable x follows "Gassian/ Normal" distribution.

```python
# Y ~ Continous Uniform Distribution(0,1)
y = np.random.uniform(0,1,10000);    # Y~U(0,1) :- Uniformly distributed.
sns.kdeplot(np.array(y), bw=0.1)
plt.show()
```
