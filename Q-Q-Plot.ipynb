# Q-Q Plot

```python
#Q-Q plot
import numpy as np 
import pylab 
import scipy.stats as stats

# N(0,1)
std_normal = np.random.normal(loc = 0, scale = 1, size=1000)
#(loc--> Mean , scale--> std_dev and size-->no.of samples)

# 0 to 100th percentiles of std-normal
for i in range(0,101):
    print(i, np.percentile(std_normal,i))


```

```python
# generate 100 samples from N(20,5) #(mean, std-dev)
measurements = np.random.normal(loc=20, scale=5, size=100)
# try size 100

stats.probplot(measurements, dist='norm', plot=pylab) # 'norm' = standatd normal variate --> N(0,1)
      # probplot(probability plot helps us to plot Q-Q:Quantile-Quantile-plot)
pylab.show()

# Here, we are comparing plot between Gaussian/Normal distribution(random.normal) and Standard normal variate('norm')
# So, since the two distributions are normal then the points lie on the straightline.
```

```python
# generate 1000 samples from N(20,5) #(mean, std-dev)
measurements = np.random.normal(loc=20, scale=5, size=1000)

stats.probplot(measurements, dist='norm', plot=pylab) # 'norm' = standatd normal variate --> N(0,1)
      # probplot(probability plot helps us to plot Q-Q:Quantile-Quantile-plot)
pylab.show()
```

```python
# generate 100 samples from N(20,5)
measurements = np.random.uniform(low=-1, high=1, size=100) # Uniform random distribution.
# Try size 1000

stats.probplot(measurements, dist='norm',plot=pylab)
pylab.show()
```

```python
# generate 1000 samples from N(20,5)
measurements = np.random.uniform(low=-1, high=1, size=10000) # Uniform random distribution.

stats.probplot(measurements, dist='norm',plot=pylab)
pylab.show()

# Here, we are comparing plot between Gaussian/Normal distribution(random.normal) on x-axis 
# and uniform distribution (random.uniform) on y-axis.
# So, since the two distributions are different then the points should not lie on straightline.
# (As we can see in the below plot they did not lie on the straight line.)
```

-> If all of the points are lie on the straight line then we can conclude that the distribution on 'x-axis' and 'y-axis' are from the same family (ex:- Gaussian/Normal)

# Notes:-

-> As the sample size increses, the more number of points will fall closer(on the) to the line.

-> Limitation of Q-Q plot:- If the no. of oberservations are small then it's hard to interpret Q-Q plot.
